## Random (x, y, radius) for bubble plot

```python
N = 100
x = np.random.rand(N)
y = np.random.rand(N)
colors = np.random.rand(N)
area = np.pi * (15 * np.random.rand(N))**2  # 0 to 15 point radiuses
plt.scatter(x, y, s=area, c=colors, alpha=0.5)
plt.savefig('fig.png')
```

## Shuffle data and plot again

```python
from random import shuffle
comb = list(zip(x,y, area, colors))
shuffle(comb)
x1, y1, area1, colors1 = zip(*comb)
plt.scatter(x1, y1, s=area1, c=colors1, alpha=0.5)
plt.savefig('fig1.png')
```

## Side by side comparison

```python
plt.figure(figsize=[12,4])
plt.subplot(1, 2, 1)
plt.scatter(x, y, s=area, c=colors, alpha=0.5)

plt.subplot(1, 2, 2)
plt.scatter(x1, y1, s=area1, c=colors1, alpha=0.5)

plt.show()
```

## Blink comparator

```python
%matplotlib nbagg
%matplotlib qt
from matplotlib import animation
bfig = plt.figure()
img = plt.imread('fig.png')
img1 = plt.imread('fig1.png')
blink = animation.ArtistAnimation(bfig, [[plt.imshow(img)], [plt.imshow(img1)]], blit=True, repeat_delay=0, interval=1000)
plt.axis('off')
plt.show()
```

## Diff rasters

```python
img = plt.imread('fig.png')
img1 = plt.imread('fig1.png')
diff = np.copy(img)
for t,v in np.ndenumerate(img):
    x,y,z = t
    v2 = img1[x][y][z]
    if v != v2:
        diff[x][y][z] = (v + v2)/2
    else:
        diff[x][y][z] = 1


plt.imshow(diff)
plt.savefig('diff.png')
```
